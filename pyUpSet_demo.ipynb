# A simple demo that creates new DataFrames instead of loading the pickled example data

```python
%pylab inline
import pyupset as pyu
import pandas as pd
```

## Create three sets, each containing a few characters

```python
df1 = pd.DataFrame(data=['A','B','C'], columns=['Items'])
df2 = pd.DataFrame(data=['B','C','D','E'], columns=['Items'])
df3 = pd.DataFrame(data=['B','F','G','H','I','J'], columns=['Items'])

pyu.plot({'Set 1':df1, 'Set 2':df2, 'Set 3':df3});
```
